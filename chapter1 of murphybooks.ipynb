```python
import numpy as np
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))


def relu(z):
    return np.maximum(0, z)


def heaviside(z):
    return z > 0


def softplus(z):
    return np.log(1 + np.exp(z))


def lrelu(z, lam=0.1):
    return np.maximum(lam * z, z)


def elu(z, alpha=1):
    return np.where(z < 0, alpha * (jnp.exp(z) - 1), z)


def elu2(z, lam=0.5):
    return np.maximum(0, z) + np.minimum(0, lam * (np.exp(z) - 1))


def swish(z):
    return z * sigmoid(z)
```

```python
fig = plt.figure()
a = np.linspace(-4, 4, 500)
y_sigmoid = sigmoid(a)
plt.plot(a, y_sigmoid, c="tab:blue")
plt.xlabel("$a$")
plt.ylabel("$\sigma(a)$")
plt.title("Sigmoid function")
sns.despine()
savefig("sigmoid_activation.pdf")
```

```python
# alpha and scale to self normalize with mean 0 and standard deviation 1
# (see equation 14 in the SELU paper):
alpha_0_1 = -jnp.sqrt(2 / jnp.pi) / (erfc(1 / jnp.sqrt(2)) * jnp.exp(1 / 2) - 1)
scale_0_1 = (
    (1 - erfc(1 / jnp.sqrt(2)) * jnp.sqrt(jnp.e))
    * jnp.sqrt(2 * jnp.pi)
    * (
        2 * erfc(jnp.sqrt(2)) * jnp.e**2
        + jnp.pi * erfc(1 / jnp.sqrt(2)) ** 2 * jnp.e
        - 2 * (2 + jnp.pi) * erfc(1 / jnp.sqrt(2)) * jnp.sqrt(jnp.e)
        + jnp.pi
        + 2
    )
    ** (-1 / 2)
)


def selu(z, scale=scale_0_1, alpha=alpha_0_1):
    return scale * elu(z, alpha)


z = jnp.linspace(-5, 5, 200)

# plt.figure(figsize=(11,4))
plt.figure()
plt.plot(z, sigmoid(z), "b-", linewidth=2, label="Sigmoid")
plt.plot(z, jnp.tanh(z), "g--", linewidth=2, label="Tanh")
plt.plot(z, relu(z), "m-.", linewidth=2, label="ReLU")
plt.grid(True)
plt.legend(bbox_to_anchor=(1.1, 1), fontsize=14)
plt.title("Activation functions", fontsize=14)
plt.axis([-5, 5, -1.2, 1.2])
plt.show()
```

```python
# plt.figure(figsize=(11,4))
plt.figure()
plt.plot(z, relu(z), "r-", linewidth=2, label="ReLU")
plt.plot(z, lrelu(z), "g--", linewidth=2, label="LReLU")
plt.plot(z, elu(z), "b-", linewidth=2, label="ELU")
plt.plot(z, selu(z), "k:", linewidth=2, label="SELU")
plt.plot(z, swish(z), "m-.", linewidth=2, label="swish")
plt.grid(True)
plt.legend(bbox_to_anchor=(.5, 1), fontsize=12)
plt.title("Activation functions", fontsize=14)
plt.axis([-2, 2, -1.2, 2])
plt.show()
```
